```python
#imports
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

### Confusion matrix for 2 class  classification problem

```python


# Actual and predicted drink types (0 for tea, 1 for coffee)
y_true = [0, 0, 1, 1, 0, 1, 0, 0, 1, 1]  # Actual
y_pred = [0, 1, 1, 1, 0, 0, 0, 1, 1, 1]  # Predicted

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Tea', 'Coffee'])
plt.title('Confusion Matrix with two class variables')
disp.plot(cmap='Blues') 
```

### Confusion matrix for 3 class  classification problem

```python

# Actual and predicted drink types (0 for tea, 1 for coffee, 2 for hot chocolate)
y_true = [0, 1, 2, 1, 0, 1, 0, 2, 1, 2]  # Actual
y_pred = [0, 2, 1, 1, 0, 0, 0, 2, 1, 2]  # Predicted
# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred)
# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Tea', 'Coffee', 'Hot Chocolate'])

disp.plot(cmap='Blues')
plt.title('Confusion Matrix with three class variables')
plt.show()
```

### Accuracy, Precision and Recall

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
# Actual and predicted drink types (0 for tea, 1 for coffee)
y_true = [0, 0, 1, 1, 0, 1, 0, 0, 1, 1]  # Actual
y_pred = [0, 1, 1, 1, 0, 0, 0, 1, 1, 1]  # Predicted
# Compute accuracy, precision, and recall
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='binary')
recall = recall_score(y_true, y_pred, average='binary')
print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
```
